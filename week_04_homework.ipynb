13.Use boolean indexing to create an array of `True/False` using `origin` and the value 'US'.

```python
#import numpy as np
filter_origin = []

for element in origin:
    if element == 'US':
        filter_origin.append(True)
    else:
        filter_origin.append(False)
    
new_origin = origin[filter_origin]
print(filter_origin)
print(new_origin)
```
